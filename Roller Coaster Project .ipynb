```python
import pandas as pd
rollers = pd.read_csv('rollercoasterdata.csv')
rollers.columns = rollers.columns.str.strip()
rollers.head()
```

```python
%load_ext pycodestyle_magic
```

```python

import time, datetime
# rank the rollercoasters by spped and give them points based on their ranks
rollers_speed = rollers.dropna(subset=["Speed (mph)"])
rollers['Speed Score'] = rollers_speed["Speed (mph)"].rank(ascending=True)
rollers_speed.sort_values(by="Speed (mph)", ascending=False)
rollers['Speed Score'].fillna(0, inplace=True)

# rank the rollercoasters by drops and give points based on their ranks
rollers_drops = rollers.dropna(subset=["Drop (feet)"])
rollers['Drop Score'] = rollers_drops["Drop (feet)"].rank(ascending=True)
rollers_drops.sort_values(by="Drop (feet)", ascending=False)
rollers['Drop Score'].fillna(0, inplace=True)


# wood bonus
def bonus(construction):
    if construction == 'wood':
        return 10
    elif construction == 'steel':
        return 0
    else:
        return 0


rollers['Construction Bonus'] = rollers.Construction.apply(bonus)

# fill in / replace the blank Durationa will 0:00
rollers['Duration (min:sec)'] = rollers['Duration (min:sec)'].fillna("0:00")
rollers['Duration (min:sec)'] = rollers['Duration'
                                        ' (min:sec)'].replace(" ", "0:00")
rollers['Duration (min:sec)'] = pd.to_datetime(rollers['Duration (min:sec)'], format='%M:%S')
rollers.head()


# convert duration time into seconds
def convert(t):
    return int(datetime.timedelta(minutes=t.minute,
                                  seconds=t.second).total_seconds())


rollers['Duration (min:sec)'] = rollers['Duration (min:sec)'].apply(convert)
rollers.head()

# rank the rollercoasters by duration and give them points based on their ranks
rollers_duration = rollers.dropna(subset=['Duration (min:sec)'])
rollers['Duration Score'] = rollers_duration['Duration'
                                             ' (min:sec)'].rank(ascending=True)
rollers_duration.sort_values(by="Duration (min:sec)", ascending=False)

rollers['Total Score'] = rollers['Duration Score'] + rollers['Speed Score']
rollers['Total Score'] = rollers['Total Score'] + rollers['Drop Score']
rollers['Total Score'] = rollers['Total Score'] + rollers['Construction Bonus']
rollers = rollers.sort_values(by="Total Score", ascending=False)
rollers[0:10]
```

```python
rollers.to_csv('newdata.csv')
```
